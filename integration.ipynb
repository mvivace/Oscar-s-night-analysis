## 2. Importiamo i dataset necessari per effettuare l'integrazione relativa ai film 

Per effettuare integrazione sono stati utilizzati diversi dataset provenienti dai siti TMDB e IMDB che forniscono informazioni aggiuntive sui film. Oltre a questi sono stati utilizzati due ulteriori dataset, uno sui codici delle lingue per poter modificare adeguatamente la colonna "language" del dataset finale e il dataset codici_imdb contenente i soli codici dei film di nostro interesse. 
I dataset estratti da TMDB sono due: "MOVIE" e "tmdb_all_data", sono stati utilizzati entrambi poiché quest'ultimo conteneva informazioni aggiuntive rispetto al primo che era meno recente. 

```python
tmdb = pd.read_csv(r"dataset/MOVIE.csv")
tmdb_all_data = pd.read_csv(r"dataset/tmdb_all_data.csv", lineterminator='\n')
imdb =  pd.read_json(r"dataset/movies_merged.json")
codiciLingue = pd.read_csv(r"dataset/codicilingue.csv")
codici_imdb = pd.read_csv(r"dataset/imdbCodici.csv")
```

Effettuiamo l'intersezione fra il dataset "codici_imdb" e il dataset "tmdb" per poter costruire un dataset con i soli film di interesse e con le sole colonne di interesse ottenendo per l'appunto il dataset "tmdb_minore1"

```python
tmdb_minore = codici_imdb.merge(tmdb, how = 'left', left_on = 'IMDB_ID', right_on = 'IMDB_ID' )
tmdb_minore1 = tmdb_minore[['TITLE', 'IMDB_ID', 'ORIGINAL_LANGUAGE', 'ORIGINAL_TITLE', 'OVERVIEW', "POPULARITY", "POSTER_PATH", "RELEASE_DATE", "RUNTIME", "VOTE_AVERAGE", 'VOTE_COUNT']]
```

```python
tmdb_all_data_ridotto = codici_imdb.merge(tmdb_all_data, how = 'left', left_on = 'IMDB_ID', right_on = 'imdb_id' )
tmdb_all_data_ridotto1 = tmdb_all_data_ridotto[['IMDB_ID', 'genres', 'original_language', 'original_title', 'overview', 'popularity', 'poster_path', 'runtime', 'revenue', 'title', 'vote_average', 'vote_count']]
```

Effettuiamo l'intersezione fra il dataset "codici_imdb" e il dataset "imdb" per poter costruire un dataset con i soli film di interesse ottenendo il dataset "imdb_minore"

```python
imdb_minore = codici_imdb.merge(imdb, how = 'left', left_on = 'IMDB_ID', right_on = 'ImdbId' )
```

Una volta ottenuti i tre dataset con i soli film di interesse, integriamo i dataset "tmdb_minore1" e "tmdb_all_data_ridotto1" utilizzando come chiave il codice imdb presente in tutti e tre i dataset e otteniamo un dataset completo di tutte le informazioni provenienti dal sito TMDB. Successivamente Integriamo quest'ultimo con il dataset "imdb_minore" che contiene ulteriori informazioni interessanti provenienti da sito IMDB. Il dataset finale prende il nome "integratoTMDBeIMDBCompleto".

```python
integratoTMDBCompleto = pd.merge(tmdb_minore1,tmdb_all_data_ridotto1[['popularity','vote_average', 'vote_count','IMDB_ID']],left_on='IMDB_ID',right_on='IMDB_ID', how='left')
integratoTMDBeIMDBCompleto = pd.merge(imdb_minore, integratoTMDBCompleto,left_on='IMDB_ID',right_on='IMDB_ID', how='left')
```

```python
integratoTMDBeIMDBCompleto
```

Poiché le colonne "cast" e "director" contengono degli annidamenti trasformiamo il dataset in un formato json in modo tale da applicare una funzione _json_normalize_ che consente, partendo dalla singola colonna di poter estrapolare i dati al suo interno, potendo così costruire due nuovi dataset "Cast" e Director". Tali dataset sono poi stati nuovamente integrati con il dataset "integratoTMDBeIMDBCompleto" 

```python
integratoTMDBeIMDBCompleto.to_json(path_or_buf='integratoTMDBeIMDBCompleto.json', orient='records')
```

```python
with open('integratoTMDBeIMDBCompleto.json') as f:
        result = json.load(f)
```

```python
cast = pd.json_normalize(result, record_path='cast', meta='ImdbId')
```

```python
director= pd.json_normalize(result, max_level=1)
director = director[['ImdbId', 'director.name', 'director.name_id']]
```

```python
integratoTMDBeIMDBCompleto_conCAST = pd.merge(integratoTMDBeIMDBCompleto, cast, left_on='IMDB_ID',right_on='ImdbId', how='left')
integratoTMDBeIMDBCompleto_conCASTeDIRECTOR =  pd.merge(integratoTMDBeIMDBCompleto_conCAST , director, left_on='IMDB_ID',right_on='ImdbId', how='left')
```

Creiamo un nuovo dataset "integratoTMDBeIMDBCompleto_conCASTeDIRECTOR1" tramite la funzione _explode_ che crea tante righe quanti sono i generi associati a un film e creiamo un dataset finale contenente le sole colonne di interesse "integrato_finale"

```python
integratoTMDBeIMDBCompleto_conCASTeDIRECTOR1 = integratoTMDBeIMDBCompleto_conCASTeDIRECTOR.explode('genre')
```

```python
integrato_finale = integratoTMDBeIMDBCompleto_conCASTeDIRECTOR1[['IMDB_ID','TITLE',
       'year', 'runtime', 'genre', 'ratingValue', 'ratingCount', 'ORIGINAL_LANGUAGE', 'POPULARITY',
       'POSTER_PATH', 'RELEASE_DATE', 'RUNTIME', 'VOTE_AVERAGE', 'VOTE_COUNT',
       'popularity', 'vote_average', 'vote_count', 'name_y',"name_id", 'director.name', 'director.name_id']]
```

Per la colonna "year" del dataset integrato finale estraiamo le quattro cifre dell'anno per via di simboli aggiuntivi anomali che compaiono sporadicamente, trasformaimo in formato datetime la colonna "year" e prendiamo l'anno. Per la colonna "runtime" estraiamo solo il numero (la colonna riporta anche la dicitura "min"). Trasformiamo la colonna "Ratingvalue" in float e trasformiamo  "RatingCount" in numero intero essendo un conteggio.

```python
integrato_finale['year'] = integrato_finale['year'].str.extract(r'(\d{4})')
integrato_finale['year'] = pd.DatetimeIndex(integrato_finale['year'])
integrato_finale['year'] = integrato_finale['year'].dt.to_period('Y')
integrato_finale['runtime'] = integrato_finale['runtime'].str.extract(r'(\d+)')
integrato_finale['runtime'] = integrato_finale['runtime'].astype(float)
integrato_finale['ratingValue'] = integrato_finale['ratingValue'].astype(float)
integrato_finale['ratingCount'] = integrato_finale['ratingCount'].str.replace(',','').astype(float)
integrato_finale['RUNTIME']=integrato_finale['RUNTIME'].astype('float64')
```

Poiché integrando i diversi dataset sono presenti più colonne riferite al tempo applichiamo la funzione tempo che prende il massimo dell'attributo "Runtime" tra i due dataset e lo inserisce nella colonna "runtime".

```python
def tempo(riga):
    cond = (riga['runtime']) != (riga['RUNTIME'])
    if cond == True:
        r1 = riga['runtime']
        r2 = riga['RUNTIME']
        c = max(r1,r2)
        return c
    else: return riga['runtime']
```

```python
integrato_finale['runtime'] = integrato_finale.apply(tempo, axis=1)
integrato_finale['RELEASE_DATE'] = pd.DatetimeIndex(integrato_finale['RELEASE_DATE'])
integrato_finale['RELEASE_DATE'] = integrato_finale['RELEASE_DATE'].dt.to_period('Y')
```

Poiché integrando i diversi dataset sono presenti più colonne riferite all'anno applichiamo la funzione "annoUscita" che prende l'anno più piccolo tra i due dataset e lo inserisce nella colonna "year"

```python
def annoUscita(riga):
    cond = (riga['year']) != (riga['RELEASE_DATE'])
    if cond == True:
        r1 = riga['year']
        r2 = riga['RELEASE_DATE']
        c = min(r1,r2)
        return c
    else: return riga['year']
```

```python
integrato_finale['year'] = integrato_finale.apply(annoUscita, axis=1)
```

La funzione lingua, trasforma la colonna "Original_Language" in cui è presente la sigla della lingua del film in nome completo della lingua originale.

```python
def lingua(riga):
    for i,j in zip(codiciLingue['sigla'], codiciLingue['language']):
        cond = (riga == i)
        if cond == True:
            return j
```

```python
integrato_finale['ORIGINAL_LANGUAGE'] = integrato_finale['ORIGINAL_LANGUAGE'].apply(lingua)
```

Rinominiamo ora le colonne in modo che risulti più chiaro e manteniamo solo quelle di interesse, ottenendo il dataset definitivo

```python
integrato_finale = integrato_finale.rename(columns={"IMDB_ID": "IMDB_Id", "TITLE": "Title", "year": "Year", "director.name": "Director", "name_y":"Cast","genre": "Genre", "runtime": "Runtime", "ORIGINAL_LANGUAGE": "OriginalLanguage", "ratingValue" : "ratingValue_IMDB", "ratingCount": "ratingCount_IMDB", "VOTE_AVERAGE": "ratingValue_TMDB0", "VOTE_COUNT":"ratingCount_TMDB0" , "POPULARITY":"Popularity_TMDB0", "vote_average":"ratingValue_TMDB1","vote_count": "ratingCount_TMDB1", "popularity" :"Popularity_TMDB1", "POSTER_PATH":"PosterPath" , "name_id": "actorId","director.name_id": "directorId"})
```

```python
integrato_finale = integrato_finale[['IMDB_Id', 'Title', 'Year', 'Director', "directorId", 'Cast', "actorId", 'Genre', 'Runtime',
       'OriginalLanguage', 'ratingValue_IMDB','ratingValue_TMDB0','ratingValue_TMDB1', 'ratingCount_IMDB',
        'ratingCount_TMDB0', 'ratingCount_TMDB1','Popularity_TMDB0', 'Popularity_TMDB1','PosterPath']]
```

## 3. Importiamo i dataset necessari per effettuare l'integrazione relativa ad attori e attrici 

L'obiettivo ora è quello di integrare al dataset precedente anche le informazioni relative ad attori/attrici e a direttori che sono presenti nel dataset "person" estrapolato anch'esso dal sito TMDB. Una volta caricato il file, manteniamo solo due colonne: "A_D_Id" e "POPULARITY". L'integrazione è resa possibile grazie alla colonna "A_D_Id", il codice identificativo di attori, attrici, direttori, direttrici che rappresenta la chiave. 

```python
person = pd.read_csv(r"dataset/PERSON1.csv")
person = person[['IMDB_ID','POPULARITY']]
```

```python
person = person.rename(columns={"IMDB_ID": "A_D_Id"})
```

Integriamo al dataset "integrato_finale" le informazioni relative ad attori/attrici e successivamente quelle relative a direttori/direttrici

```python
integrato_attore = integrato_finale.merge(person, how='left', left_on=['actorId'], right_on=["A_D_Id"])
integrato = integrato_attore.merge(person, how='left', left_on=['directorId'], right_on=["A_D_Id"], suffixes=('_Actor','_Director'))
```

```python
integrato.to_csv(r"results/integrated.csv")
```
